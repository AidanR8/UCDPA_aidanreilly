```python
#making subsets of each df for the time period we want 2004-2021
DonegalMonthlyData = DonegalMonthlydf.iloc[584:800 , 0:13].copy()
WestmeathMonthlyData = WestmeathMonthlydf.iloc[648:864 , 0:13].copy()
MayoMonthlyData = MayoMonthlydf.iloc[88:303 , 0:13].copy()
CorkMonthlyData = CorkMonthlydf.iloc[504:719 , 0:13].copy()
DublinMonthlyData = DublinMonthlydf.iloc[746:961 , 0:13].copy()
```

```python
#code to replace missing NaN values with rando number between min and max in column

numeric_cols = DonegalMonthlyData.select_dtypes([np.number]).columns

DonegalMonthlyData[numeric_cols] = DonegalMonthlyData[numeric_cols].apply(lambda x: x.fillna(pd.Series(np.random.uniform(x.min(), x.max(), len(x)))))
```
