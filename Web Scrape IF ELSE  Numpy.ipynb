```python
# IF ELSE STATEMENT EXAMPLE

a = 25
b = 50

if a > b:
    print('a is greater than b')
else:
    print('a is less than b')
```

```python
# NUMPY - example of creating a numpy array and using for and .flat to create a new flattened version of this array

import numpy as np

array = np.array([[2, 4, 6,], 
                 [8, 10, 12], 
                 [14,16,18]
                 ])  
                 
print(array)



for x in array.flat:
    print(x)
```

```python
# example of linking two arrays with numpy

a = np.array([[2, 4, 6,],
              [8, 10, 12],
              [14,16,18]
              ])  


b = np.array([[1, 3, 5,],
              [7, 9, 11],
              [13,15,17]
              ])  
                 
c = np.concatenate((a,b))

print(c)
```
