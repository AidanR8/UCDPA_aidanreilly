```python
import pandas as pd

DonegalMonthlydf = pd.read_csv('C:/Users./BYO User/UCD Data Project/Downloads for Data Project/DonegalMonthly.csv', usecols= ["year", "month", "rain", "maxtp", "mintp", "wdsp", "sun"], skiprows=19, sep = '\s*,\s*', engine = 'python')
DonegalMonthlydf.head() #and add this to see if our code worked
```

```python
# new df from original df and drop year column as we have already identified
DonegalMonthly2001 = DonegalMonthlydf.iloc[548:560 , 0:13].copy()
DonegalMonthly2006 = DonegalMonthlydf.iloc[608:620 , 0:13].copy()
DonegalMonthly2011 = DonegalMonthlydf.iloc[668:680 , 0:13].copy()
DonegalMonthly2017 = DonegalMonthlydf.iloc[740:752 , 0:13].copy()
DonegalMonthly2021 = DonegalMonthlydf.iloc[788:800 , 0:13].copy()
```

```python
import calendar
DonegalMonthly2001['month'] = DonegalMonthly2001['month'].apply(lambda x: calendar.month_abbr[x])
DonegalMonthly2006['month'] = DonegalMonthly2006['month'].apply(lambda x: calendar.month_abbr[x])
DonegalMonthly2011['month'] = DonegalMonthly2011['month'].apply(lambda x: calendar.month_abbr[x])
DonegalMonthly2017['month'] = DonegalMonthly2017['month'].apply(lambda x: calendar.month_abbr[x])
DonegalMonthly2021['month'] = DonegalMonthly2021['month'].apply(lambda x: calendar.month_abbr[x])
```

```python
# identify MIN and MAX in original DonegalMonthlydf for the sun column

Donegal_gmin_min11 = DonegalMonthlydf['sun'].loc[DonegalMonthlydf['sun'].idxmin()]
Donegal_gmin_max11 = DonegalMonthlydf['sun'].loc[DonegalMonthlydf['sun'].idxmax()]
print(Donegal_gmin_min11)
print(Donegal_gmin_max11)
```

```python
import numpy as np

# code to replace NaN values with random number between the min and max range we have identified for sun in entire original Donegal weather data

DonegalMonthly2011['sun'] = DonegalMonthly2011['sun'].mask(DonegalMonthly2011['sun'].isna(), np.random.uniform(9.5, 287.5, size=len(DonegalMonthly2011)))

# use .loc to show row 829 which we identified from above and see that it is now updated with a random number from our defined min/max range and no longer showing as NaN

DonegalMonthly2011.loc[[668]]
```

```python
# amending the missing values has caused the float number of figures after decimal point to expand
# change decimal point back to 1 place and view again toensure decimal places have reverted to what we want e.g. 1 place

pd.set_option('display.float_format', '{:.1f}'.format)
DonegalMonthly2011
```
