```python
# creating a new DF subset as DonegalMonthlyData using iloc which we will then begin to manipulate

MayoMonthlyData = MayoMonthlydf.iloc[88:303 , 0:13].copy()
MayoMonthlyData
```

```python
# adding a new column of data to show the location identifier for each row of data to help when merging

MayoMonthlyData['Location'] = "Mayo"
MayoMonthlyData
```

```python
# use .info to show the newly created dataframe information and datatype - note reduction in entries

MayoMonthlyData.info()
```

```python
# testing to see if I can amend month numbers to actual month name abbreviated e.g. Jan, Feb, etc., using the .mask function

MayoMonthlyData['month'].mask(MayoMonthlyData['month'] ==1 ,'Jan', inplace=True)
MayoMonthlyData

# this is a success
```

```python
# displaying to show that DonegalMonthlydf has not been affected and that my new dataframe is being manipulated
# original dataframe data remains the same

MayoMonthlydf
```

```python
# amending all months from number to a descriptor using .mask

MayoMonthlyData['month'].mask(MayoMonthlyData['month'] ==1 ,'Jan', inplace=True)
MayoMonthlyData['month'].mask(MayoMonthlyData['month'] ==2 ,'Feb', inplace=True)
MayoMonthlyData['month'].mask(MayoMonthlyData['month'] ==3 ,'Mar', inplace=True)
MayoMonthlyData['month'].mask(MayoMonthlyData['month'] ==4 ,'Apr', inplace=True)                              
MayoMonthlyData['month'].mask(MayoMonthlyData['month'] ==5 ,'May', inplace=True)                              
MayoMonthlyData['month'].mask(MayoMonthlyData['month'] ==6 ,'Jun', inplace=True)
MayoMonthlyData['month'].mask(MayoMonthlyData['month'] ==7 ,'Jul', inplace=True)
MayoMonthlyData['month'].mask(MayoMonthlyData['month'] ==8 ,'Aug', inplace=True)
MayoMonthlyData['month'].mask(MayoMonthlyData['month'] ==9 ,'Sep', inplace=True)
MayoMonthlyData['month'].mask(MayoMonthlyData['month'] ==10 ,'Oct', inplace=True)
MayoMonthlyData['month'].mask(MayoMonthlyData['month'] ==11 ,'Nov', inplace=True)
MayoMonthlyData['month'].mask(MayoMonthlyData['month'] ==12 ,'Dec', inplace=True)
MayoMonthlyData
```

```python
# identify missing values in the new dataframe and sum

MayoMonthlyData.isna().sum()
```

```python
# locate the missing values for pre change review using .loc, .isnull and .any with axis set to 1 to perform on columns that satisy condition 

MayoMonthlyData.loc[MayoMonthlyData.isnull().any(axis=1)]
```

```python
# identify MIN and MAX in mintp column with NaN for mintp

Mayo_mintp_min = MayoMonthlyData['mintp'].loc[MayoMonthlyData['mintp'].idxmin()]
Mayo_mintp_max = MayoMonthlyData['mintp'].loc[MayoMonthlyData['mintp'].idxmax()]
print(Mayo_mintp_min)
print(Mayo_mintp_max)
```

```python
# code to replace NaN values with random number between the min and max range we have identified in mintp
MayoMonthlyData['mintp'] = MayoMonthlyData['mintp'].mask(MayoMonthlyData['mintp'].isna(), np.random.uniform(-7.4, 9.8, size=len(MayoMonthlyData)))

# use .loc to show a row e.g. 98 which we identified from above and see that it is now updated with a random number from our defined min/max range and no longer showing as NaN

MayoMonthlyData.loc[[98]]
```

```python
# rerun our code we used to locate missing values (mnmin, gmin and sun)

Mayo_mnmin_min = MayoMonthlyData['mnmin'].loc[MayoMonthlyData['mnmin'].idxmin()]
Mayo_mnmin_max = MayoMonthlyData['mnmin'].loc[MayoMonthlyData['mnmin'].idxmax()]
Mayo_gmin_min = MayoMonthlyData['gmin'].loc[MayoMonthlyData['gmin'].idxmin()]
Mayo_gmin_max = MayoMonthlyData['gmin'].loc[MayoMonthlyData['gmin'].idxmax()]
Mayo_sun_min = MayoMonthlyData['sun'].loc[MayoMonthlyData['sun'].idxmin()]
Mayo_sun_max = MayoMonthlyData['sun'].loc[MayoMonthlyData['sun'].idxmax()]

print(Mayo_mnmin_min)
print(Mayo_mnmin_max)
print(Mayo_gmin_min)
print(Mayo_gmin_max)
print(Mayo_sun_min)
print(Mayo_sun_max)
```

```python
# code to replace NaN values with random number between the min and max range we have identified in mnmin, gmin and sun columns

MayoMonthlyData['mnmin'] = MayoMonthlyData['mnmin'].mask(MayoMonthlyData['mnmin'].isna(), np.random.uniform(-1.5, 12.5, size=len(MayoMonthlyData)))
MayoMonthlyData['gmin'] = MayoMonthlyData['gmin'].mask(MayoMonthlyData['gmin'].isna(), np.random.uniform(-13.6, 7.4, size=len(MayoMonthlyData)))
MayoMonthlyData['sun'] = MayoMonthlyData['sun'].mask(MayoMonthlyData['sun'].isna(), np.random.uniform(15.5, 223.0, size=len(MayoMonthlyData)))
```

```python
# we can now confirm that all NaN values have been replaced by running the .isna function again

MayoMonthlyData.isna().sum()
```
