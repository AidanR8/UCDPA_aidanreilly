```python
# creating a new DF subset as CorkMonthlyData using iloc which we will then begin to manipulate

CorkMonthlyData = CorkMonthlydf.iloc[504:719 , 0:13].copy()
CorkMonthlyData
```

```python
# adding a new column of data to show the location identifier for each row of data to help when merging

CorkMonthlyData['Location'] = "Cork"
CorkMonthlyData
```

```python
# use .info to show the newly created dataframe information and datatype - note reduction in entries

CorkMonthlyData.info()
```

```python
# testing to see if I can amend month numbers to actual month name abbreviated e.g. Jan, Feb, etc., using the .mask function

CorkMonthlyData['month'].mask(CorkMonthlyData['month'] ==1 ,'Jan', inplace=True)
CorkMonthlyData

# this is a success
```

```python
# displaying to show that CorkMonthlydf has not been affected and that my new dataframe is being manipulated
# original dataframe data remains the same

CorkMonthlydf
```

```python
# amending all months from number to a descriptor using .mask

CorkMonthlyData['month'].mask(CorkMonthlyData['month'] ==1 ,'Jan', inplace=True)
CorkMonthlyData['month'].mask(CorkMonthlyData['month'] ==2 ,'Feb', inplace=True)
CorkMonthlyData['month'].mask(CorkMonthlyData['month'] ==3 ,'Mar', inplace=True)
CorkMonthlyData['month'].mask(CorkMonthlyData['month'] ==4 ,'Apr', inplace=True)                              
CorkMonthlyData['month'].mask(CorkMonthlyData['month'] ==5 ,'May', inplace=True)                              
CorkMonthlyData['month'].mask(CorkMonthlyData['month'] ==6 ,'Jun', inplace=True)
CorkMonthlyData['month'].mask(CorkMonthlyData['month'] ==7 ,'Jul', inplace=True)
CorkMonthlyData['month'].mask(CorkMonthlyData['month'] ==8 ,'Aug', inplace=True)
CorkMonthlyData['month'].mask(CorkMonthlyData['month'] ==9 ,'Sep', inplace=True)
CorkMonthlyData['month'].mask(CorkMonthlyData['month'] ==10 ,'Oct', inplace=True)
CorkMonthlyData['month'].mask(CorkMonthlyData['month'] ==11 ,'Nov', inplace=True)
CorkMonthlyData['month'].mask(CorkMonthlyData['month'] ==12 ,'Dec', inplace=True)
CorkMonthlyData
```

```python
# identify missing values in the new dataframe and sum

CorkMonthlyData.isna().sum()
```

```python
#locate the missing mintp and gmin values for pre change review using .loc, .isnull and .any with axis set to 1 to perform on columns that satisy condition 

CorkMonthlyData.loc[CorkMonthlyData.isnull().any(axis=1)]
```

```python
# identify MIN and MAX in gmin column with NaN for gmin

Cork_gmin_min = CorkMonthlyData['gmin'].loc[CorkMonthlyData['gmin'].idxmin()]
Cork_gmin_max = CorkMonthlyData['gmin'].loc[CorkMonthlyData['gmin'].idxmax()]
print(Cork_gmin_min)
print(Cork_gmin_max)
```

```python
# identify MIN and MAX in mintp column

Cork_mintp_min = CorkMonthlyData['mintp'].loc[CorkMonthlyData['mintp'].idxmin()]
Cork_mintp_max = CorkMonthlyData['mintp'].loc[CorkMonthlyData['mintp'].idxmax()]
print(Cork_mintp_min)
print(Cork_mintp_max)
```

```python
# code to replace NaN values with random number between the min and max range we have identified in gmin and mintp
CorkMonthlyData['gmin'] = CorkMonthlyData['gmin'].mask(CorkMonthlyData['gmin'].isna(), np.random.uniform(-10.3, 6.9, size=len(CorkMonthlyData)))
CorkMonthlyData['mintp'] = CorkMonthlyData['mintp'].mask(CorkMonthlyData['mintp'].isna(), np.random.uniform(-7.2, 9.9, size=len(CorkMonthlyData)))

# use .loc to show the rows in question now updated with a random numbers from our defined min/max range and no longer showing as NaN

CorkMonthlyData.loc[[504,549,592,610]]
```

```python
# amending the missing values has caused the float number of figures after decimal point to expand
# change decimal point back to 1 place and view again toensure decimal places have reverted to what we want e.g. 1 place
pd.set_option('display.float_format', '{:.1f}'.format)
CorkMonthlyData
```

```python
# rerun our code we used to locate missing values (mintp and gmin) for post change review to see what it returns
# as expected now showing nothing as all NaN values replaced

CorkMonthlyData.loc[CorkMonthlyData.isnull().any(axis=1)]
```

```python
# this can be further confirmed by running the .isna function again - all missing values have been replaced with values

CorkMonthlyData.isna().sum()
```
