```python
# testing to see if I can amend month numbers to actual month name abbreviated e.g. Jan, Feb, etc., using the .mask function

DonegalMonthlyData['month'].mask(DonegalMonthlyData['month'] ==1 ,'Jan', inplace=True)
DonegalMonthlyData

# this is a success
```

```python
# displaying to show that DonegalMonthlydf has not been affected and that my new dataframe is being manipulated
# original dataframe data remains the same

DonegalMonthlydf
```

```python
# amending all months from number to a descriptor using .mask

DonegalMonthlyData['month'].mask(DonegalMonthlyData['month'] ==1 ,'Jan', inplace=True)
DonegalMonthlyData['month'].mask(DonegalMonthlyData['month'] ==2 ,'Feb', inplace=True)
DonegalMonthlyData['month'].mask(DonegalMonthlyData['month'] ==3 ,'Mar', inplace=True)
DonegalMonthlyData['month'].mask(DonegalMonthlyData['month'] ==4 ,'Apr', inplace=True)                              
DonegalMonthlyData['month'].mask(DonegalMonthlyData['month'] ==5 ,'May', inplace=True)                              
DonegalMonthlyData['month'].mask(DonegalMonthlyData['month'] ==6 ,'Jun', inplace=True)
DonegalMonthlyData['month'].mask(DonegalMonthlyData['month'] ==7 ,'Jul', inplace=True)
DonegalMonthlyData['month'].mask(DonegalMonthlyData['month'] ==8 ,'Aug', inplace=True)
DonegalMonthlyData['month'].mask(DonegalMonthlyData['month'] ==9 ,'Sep', inplace=True)
DonegalMonthlyData['month'].mask(DonegalMonthlyData['month'] ==10 ,'Oct', inplace=True)
DonegalMonthlyData['month'].mask(DonegalMonthlyData['month'] ==11 ,'Nov', inplace=True)
DonegalMonthlyData['month'].mask(DonegalMonthlyData['month'] ==12 ,'Dec', inplace=True)
DonegalMonthlyData
```

```python
# identify missing values in the new dataframe and sum

DonegalMonthlyData.isna().sum()
```

```python
# locate the missing values for pre change review using .loc, .isnull and .any with axis set to 1 to perform on columns that satisy condition 
# too many to show below

DonegalMonthlyData.loc[DonegalMonthlyData.isnull().any(axis=1)]
```

```python
# identify MIN and MAX in mintp

Donegal_mintp_min = DonegalMonthlyData['mintp'].loc[DonegalMonthlyData['mintp'].idxmin()]
Donegal_mintp_max = DonegalMonthlyData['mintp'].loc[DonegalMonthlyData['mintp'].idxmax()]
print(Donegal_mintp_min)
print(Donegal_mintp_max)
```

```python
# code to replace NaN values with random number between the min and max range we have identified in mintp
DonegalMonthlyData['mintp'] = DonegalMonthlyData['mintp'].mask(DonegalMonthlyData['mintp'].isna(), np.random.uniform(-3.6, 11.1, size=len(DonegalMonthlyData)))

# use .loc to show the row in question now updated with a random number from our defined min/max range and no longer showing as NaN

DonegalMonthlyData.loc[[619]]
```

```python
# rerun our code we used to locate missing values (wdsp and sun)

Donegal_wdsp_min = DonegalMonthlyData['wdsp'].loc[DonegalMonthlyData['wdsp'].idxmin()]
Donegal_wdsp_max = DonegalMonthlyData['wdsp'].loc[DonegalMonthlyData['wdsp'].idxmax()]
Donegal_sun_min = DonegalMonthlyData['sun'].loc[DonegalMonthlyData['sun'].idxmin()]
Donegal_sun_max = DonegalMonthlyData['sun'].loc[DonegalMonthlyData['sun'].idxmax()]

print(Donegal_wdsp_min)
print(Donegal_wdsp_max)
print(Donegal_sun_min)
print(Donegal_sun_max)
```

```python
# code to replace NaN values with random number between the min and max range we have identified in wdsp and sun columns

DonegalMonthlyData['wdsp'] = DonegalMonthlyData['wdsp'].mask(DonegalMonthlyData['wdsp'].isna(), np.random.uniform(7.8, 22.3, size=len(DonegalMonthlyData)))
DonegalMonthlyData['sun'] = DonegalMonthlyData['sun'].mask(DonegalMonthlyData['sun'].isna(), np.random.uniform(14.9, 280.8, size=len(DonegalMonthlyData)))
```

```python
# we can now confirm that all NaN values have been replaced by running the .isna function again

DonegalMonthlyData.isna().sum()
```
