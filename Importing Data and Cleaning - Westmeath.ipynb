```python
# amending all months from number to a descriptor using .mask

WestmeathMonthlyData['month'].mask(WestmeathMonthlyData['month'] ==1 ,'Jan', inplace=True)
WestmeathMonthlyData['month'].mask(WestmeathMonthlyData['month'] ==2 ,'Feb', inplace=True)
WestmeathMonthlyData['month'].mask(WestmeathMonthlyData['month'] ==3 ,'Mar', inplace=True)
WestmeathMonthlyData['month'].mask(WestmeathMonthlyData['month'] ==4 ,'Apr', inplace=True)                              
WestmeathMonthlyData['month'].mask(WestmeathMonthlyData['month'] ==5 ,'May', inplace=True)                              
WestmeathMonthlyData['month'].mask(WestmeathMonthlyData['month'] ==6 ,'Jun', inplace=True)
WestmeathMonthlyData['month'].mask(WestmeathMonthlyData['month'] ==7 ,'Jul', inplace=True)
WestmeathMonthlyData['month'].mask(WestmeathMonthlyData['month'] ==8 ,'Aug', inplace=True)
WestmeathMonthlyData['month'].mask(WestmeathMonthlyData['month'] ==9 ,'Sep', inplace=True)
WestmeathMonthlyData['month'].mask(WestmeathMonthlyData['month'] ==10 ,'Oct', inplace=True)
WestmeathMonthlyData['month'].mask(WestmeathMonthlyData['month'] ==11 ,'Nov', inplace=True)
WestmeathMonthlyData['month'].mask(WestmeathMonthlyData['month'] ==12 ,'Dec', inplace=True)
WestmeathMonthlyData
```

```python
# identify missing values in the new dataframe and sum

WestmeathMonthlyData.isna().sum()
```

```python
# locate the missing values for pre change review using .loc, .isnull and .any with axis set to 1 to perform on columns that satisy condition 

WestmeathMonthlyData.loc[WestmeathMonthlyData.isnull().any(axis=1)]
```

```python
# identify MIN and MAX in gmin and sun columns

Westmeath_gmin_min = WestmeathMonthlyData['gmin'].loc[WestmeathMonthlyData['gmin'].idxmin()]
Westmeath_gmin_max = WestmeathMonthlyData['gmin'].loc[WestmeathMonthlyData['gmin'].idxmax()]
Westmeath_sun_min = WestmeathMonthlyData['sun'].loc[WestmeathMonthlyData['sun'].idxmin()]
Westmeath_sun_max = WestmeathMonthlyData['sun'].loc[WestmeathMonthlyData['sun'].idxmax()]

print(Westmeath_gmin_min)
print(Westmeath_gmin_max)
print(Westmeath_sun_min)
print(Westmeath_sun_max)
```

```python
# code to replace NaN values with random number between the min and max range we have identified in gmin and sun
WestmeathMonthlyData['gmin'] = WestmeathMonthlyData['gmin'].mask(WestmeathMonthlyData['gmin'].isna(), np.random.uniform(-16.6, 5.2, size=len(WestmeathMonthlyData)))
WestmeathMonthlyData['sun'] = WestmeathMonthlyData['sun'].mask(WestmeathMonthlyData['sun'].isna(), np.random.uniform(49.5, 232.4, size=len(WestmeathMonthlyData)))

# use .loc to show a row e.g. 684 which we identified from above and see that it is now updated with a random number from our defined min/max range and no longer showing as NaN

WestmeathMonthlyData.loc[[684]]
```

```python
# we can now confirm that all NaN values have been replaced by running the .isna function again

WestmeathMonthlyData.isna().sum()
```
